```python
recommend_df = df.pivot_table(index='User.ID',columns="Book.Title",values ='Book.Rating').reset_index(drop=True)
```

```python
recommend_df   #created new data frame wtih the help of pivot table which will help in finding cosine similarity.
```

```python
recommend_df = recommend_df.fillna(0)  #replacing null values with 0
```

```python
recommend_df
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim = 1-pairwise_distances(recommend_df.values,metric='cosine') #we have subtracted distances from 1 as distance is inversely proportional to similarity.
```

```python
user_sim
```

```python
user_sim = pd.DataFrame(user_sim)
```

```python
user_sim
```
